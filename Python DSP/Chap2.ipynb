```python
import os
import thinkdsp
```

```python
signalTri=thinkdsp.TriangleSignal(200,1,0)
```

```python
waveTri=signalTri.make_wave(duration=0.5, start=0, framerate=10000)
```

```python
signalTri.plot()
```

```python
spectrumTri=waveTri.make_spectrum()
spectrumTri.plot()
```

```python
spectrumTri.peaks()[:10]
```

```python
signalSquare=thinkdsp.SquareSignal(100,1,0)
```

```python
waveSquare=signalSquare.make_wave(duration=0.5,start=0, framerate=10000)
```

```python
signalSquare.plot()
```

```python
waveSquare.plot()
```

```python
waveSquare.make_spectrum().plot()
waveSquare.make_spectrum().peaks()[:30]
```

```python
framerate=10000
cos4500=thinkdsp.CosSignal(4500)
cos4500Duration=cos4500.period*5
cos4500Segment=cos4500.make_wave(cos4500Duration,framerate=10000)
cos4500Segment.plot()
cos5500Segment=cos4500.make_wave(cos4500Duration,framerate=5000000)
cos5500Segment.plot()
```

```python
cos5500=thinkdsp.CosSignal(4500)
cos5500Segment=cos5500.make_wave(cos4500Duration,framerate=10000)
cos5500Segment.plot()
cos5500Segment=cos5500.make_wave(cos4500Duration,framerate=5000000)
cos5500Segment.plot()
```
